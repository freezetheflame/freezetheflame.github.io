```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
def compute_times(lengths):
  naive_times = []
  vectorized_times = []
  for l in lengths:
    a = [i for i in range(l)];
    b = [i for i in range(l)];

    tic = time.clock()
    dot = 0.0;
    for i in range(len(a)):
        dot += a[i] * b[i]
    toc = time.clock()
    naive_times.append(toc-tic)

    n_tic = time.clock()
    n_dot_product = np.array(a).dot(np.array(b))
    n_toc = time.clock()
    vectorized_times.append(n_toc-n_tic)
  return naive_times, vectorized_times
```

```python
lengths = [10**i for i in range(7)]
naive_times, vectorized_times = compute_times(lengths)
```

```python
fig, ax = plt.subplots()

line1, = ax.plot(lengths, naive_times, label='Times with naive multiplication')
line2, = ax.plot(lengths, vectorized_times, label='Times with vectorization')
plt.title('Comparison of direct multiplication vs vectorization')
ax.legend()
plt.show()
```
